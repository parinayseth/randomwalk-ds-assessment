```python
import pandas as pd
```

```python
df = pd.read_csv('books.csv')
```

```python
df.sample(5)
```

```python
rows = df.shape[0]
columns = df.shape[1]
print(f'There are {rows} rows and {columns} columns in the book dataset')
```

```python
# checking for Null values
df.isnull().sum() 
```

```python
# Get the list of columns with null values and their counts
null_counts = df.isnull().sum()
columns_with_null = null_counts[null_counts > 0].index.tolist()

# Display data types and counts of null values for columns with null values
for column in columns_with_null:
    print(f"{column}: {df[column].dtype} - Null Count: {null_counts[column]}")
```

```python
# Printing sample for each column which has null values
df[df['isbn'].isnull()].sample(5)
```

```python
df[df['isbn13'].isnull()].sample(5)
```

```python
df[df['original_title'].isnull()].sample(5)
```

```python
df[df['language_code'].isnull()]['language_code'].sample(5)
```

```python
null_language_code_rows = df[df['language_code'].isnull()].sample(5)

# Identify the index of the 'language_code' column
language_code_index = df.columns.get_loc('language_code')

# Extract the two columns before and after 'language_code'
selected_columns = df.columns[max(0, language_code_index - 2):min(language_code_index + 3, len(df.columns))]

# Display the selected columns for the sampled rows
null_language_code_rows[selected_columns]
```
